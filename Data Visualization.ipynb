# Data Visualization

Visualisasi data bertujuan untuk mempresentasikan data dengan grafik atau bagan untuk menampilkan info yang tersembunyi didalam data. Data visual sangatlah penting dikarenakan oleh data yang ditampilkan secara visual memungkinkan banyak orang untuk memahami data dengan lebih cepat. 

Beberapa library di python untuk visualisasi data adalah:
1. Matplotlib
2. Seaborn
3. Bokeh
4. Plotly
5. ggplot

## Data Visualization dengan Matplotlib

Matplotlib adalah library di python yang digunakan untuk plotting grafik 2 dimensi dalam berbagai format. Sedangkan pyplot adalah modul untuk plotting sederhana dengan interface.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
```

### Line Plot

```python
#membuat data
x = np.arange(0,10) 
y = x ^ 2 
```

```python
plt.plot(x,y)
plt.show()
```

```python
#memberi label 
plt.plot(x,y)
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('Simple Plot') 
plt.show() 
```

```python
#mengubah warna line
plt.plot(x,y,'r')
```

```python
#membuat lebih dari 1 line
year = [1960, 1970, 1980, 1990, 2000, 2010]
l1 = [44, 58, 78, 107, 138, 170]
l2 = [50, 67, 86, 130, 153, 224]
plt.plot(year, l1,'g')
plt.plot(year, l2, 'b')
#memberi nama axis dan judul
plt.xlabel('tahun')
plt.ylabel('jumlah')
plt.title('Perbandingan jumlah dan tahun')
#memberikan keterangan
plt.annotate(xy=[1960,44], s='Data pertama line 1') 
plt.annotate(xy=[2010,224], s='Data terakhir line 2')
#memberikan lagenda
plt.legend(['Line 1','Line 2'], loc=4) 
#menyimpan plot
plt.savefig('lineplot.png',dpi=300)
#menampilkan plot
plt.show()
```

```python
#membuat beberapa plot dalam 1 gambar
fig, axs = plt.subplots(2)
fig.suptitle('Subplots Vertikal')
axs[0].plot(x, y)
axs[1].plot(x, -y)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Subplots Horizontal')
ax1.plot(x, y)
ax2.plot(x, -y)
```

### Histogram

```python
mu, sigma = 2, 0.5
v = np.random.normal(mu,sigma,10000)
```

```python
plt.hist(v, bins=70, rwidth=0.5,alpha=0.2,color='blue',edgecolor='red')
plt.xlabel('X-axis') 
plt.ylabel('Y-axis') 
plt.title('Histogram') 
plt.grid(True)
plt.show()
```

### Bar Chart

```python
left = [1, 2, 3, 4, 5] 
height = [10, 24, 36, 40, 5] 
#memberi label
tick_label = ['one', 'two', 'three', 'four', 'five'] 
#membuat bar chart
plt.bar(left, height, tick_label = tick_label, width = 0.8, color = ['red', 'green']) 
#memberi nama x-axis
plt.xlabel('x - axis') 
#memberi nama y-axis 
plt.ylabel('y - axis') 
#memberi judul plot 
plt.title('My bar chart') 
#menampilkan plot 
plt.show() 
```

```python
df=pd.DataFrame(np.random.rand(6,4),
            index=['one','two','three','four','five','six'],
            columns=pd.Index(['A','B','C','D'], name='Genus'))
df
```

```python
df.plot(kind='bar')
```

```python
df.plot(kind='bar',stacked=True,alpha=1)
```

```python
df.plot(kind='barh',stacked=True, alpha=0.5)
```

### Pie Chart

```python
#inisialisasi label
activities = ['eat', 'sleep', 'work', 'play'] 
#proporsi setiap label 
slices = [3, 7, 8, 6] 
#warna setiap label 
colors = ['r', 'y', 'g', 'b'] 
#membuat pie chart 
plt.pie(slices, labels = activities, colors=colors, 
    startangle=10, shadow = True, explode = (0, 0, 0.1, 0), 
    radius = 1.2, autopct = '%1.1f%%') 
#membuat legenda 
plt.legend() 
#menampilkan plot 
plt.show() 
```

### Boxplot

```python
#membuat data
spread = np.random.rand(50) * 100
center = np.ones(25) * 50
flier_high = np.random.rand(10) * 100 + 100
flier_low = np.random.rand(10) * -100
d1 = np.concatenate((spread, center, flier_high, flier_low), 0)
```

```python
plt.boxplot(d1)
plt.show()
```

```python
plt.figure()
plt.boxplot(d1, 0, 'rs', 0)
plt.show()
```

```python
plt.figure()
plt.boxplot(d1, 0, 'gD')
plt.show()
```

```python
#membuat data lain
d2 = np.concatenate((spread, center, flier_high, flier_low), 0)
d1.shape = (-1, 1)
d2.shape = (-1, 1)
d3 = d2[::2, 0]
data = [d1, d2, d3]
plt.figure()
plt.boxplot(data)
plt.show()
```

### Scatter Plots

```python
#membuat data
np.random.seed(19680801)
N = 50
k = np.random.rand(N)
l = np.random.rand(N)
```

```python
plt.scatter(k, l, marker='o', color='pink',label='point')
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('My Scatterplot') 
plt.legend(loc='upper left')
plt.show() 
```

### Bubble Chart

```python
m = np.random.rand(70)
n = np.random.rand(70)
o = np.random.rand(70)
colors = np.random.rand(70) 
plt.scatter(m, n, s=o*1000, c=colors, alpha=0.5)
plt.xlabel('x - axis') 
plt.ylabel('y - axis') 
plt.title('My Bubble Chart') 
plt.show()
```

### Plot Area

```python
days= [1,2,3,4,5]

sleeping= [7,8,6,11,7]
eating= [2,3,4,3,2]
working= [7,8,7,2,2]
playing= [8,5,7,8,13]
```

```python
plt.plot([],[],color='m',label='sleeping',linewidth=5)
plt.plot([],[],color='c',label='eating',linewidth=5)
plt.plot([],[],color='r',label='working',linewidth=5)
plt.plot([],[],color='k',label='playing',linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing, colors=['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot Area')
plt.legend()
plt.show()
```
